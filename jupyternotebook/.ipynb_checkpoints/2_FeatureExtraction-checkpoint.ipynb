```python
import numpy as np
import os
import librosa
import sed_eval
import scipy
```

```python
def PCNN(y,sr,parameters):
    '''
    Calculate the
        MFCCs : Mel-frequency cepstral coefficients (MFCCs)
        PCENs : using PCEN() to replace the log amplitude (dB) scaling on Mel spectra
    '''
    win_size = parameters['win_size']
    hop_size = parameters['hop_size']
    n_mels = parameters['num_mel_filters']
    n_dct = parameters['n_dct']
    fmin = parameters['min_freq']
    fmax = parameters['max_freq']
    mel_spectrogram = librosa.feature.melspectrogram(y=y, 
                                                    sr=sr,
                                                    n_fft=win_size, 
                                                    hop_length=hop_size, 
                                                    power=1, 
                                                    n_mels= n_mels,
                                                    fmin = fmin,
                                                    fmax = fmax)
    
    S_PCNN = librosa.pcen(mel_spectrogram)
    
    return scipy.fftpack.dct(S_PCNN, axis=0, type=2, norm='ortho')[:n_dct]
#=================================================================================================================
def FeatureExtraction(Files_Dir,Parameters,Property):
    #Process all the .wav, .txt, in the assigned folder
    
    #Files_Dir: Target folder for saving all raw audio data
    #Parameters: Using  for processing audio files
    #Property:Create train/test set?
    #Extract the parameters for processing audio files
    win_size = Parameters['win_size']
    hop_size = Parameters['hop_size']
    sr = Parameters['downsampling']

    file_names = []
    data = []
    labels = []
    #Calculate the Transition Matrix From Training set,saving the probability of each status transition
    SS = []
    NN = []
    SN = []
    NS = []

    file_index = 0
    for file in os.listdir(Files_Dir):
            
        filename = os.path.splitext(file)[0]
        if file.endswith('.txt'):
            print('txt file is: {}'.format(file))
            file_path = Files_Dir + '/' + file
            processed_filename =  os.path.splitext(file)[0]
            #Load Annotated Information
            annotated_event = sed_eval.io.load_event_list(file_path)
            speech_event = sed_eval.util.event_list.filter_event_list(annotated_event, scene_label=None, event_label='speech', filename=None)
            #Num of Annotated Acoustic Events
            event_num = len(annotated_event)
                
#        if file.endswith('.wav'): 
#            print(file)
#            #Check the name with .txt file
#            print(processed_filename)
#            print(os.path.splitext(file)[0])
#            if processed_filename != os.path.splitext(file)[0]:
#                raise ValueError('\nCant find the corresponding audio file to the processed .txt annotation')   
            file_index += 1
            file_names.append(file)
            #Load Audio(Dowmsampling & Force to Mono)
            audio_file = os.path.splitext(file)[0] + '.wav'
            print('wav file is: {}'.format(audio_file))
            file_path = Files_Dir + '/' + audio_file 
            audio_data,audio_sr = librosa.load(file_path,sr = sr, mono = True)
            audio_mag = np.abs(librosa.stft(audio_data, n_fft=win_size, hop_length=hop_size))  
                
            #Annotation Processing=================================================
            #list for saving number of frame labels.
            ss_temp = 0
            nn_temp = 0
            sn_temp = 0
            ns_temp = 0
            previous_s = 0
            previous_n = 0
            frame_labels_vec = np.zeros(np.ceil(len(audio_data)/hop_size).astype(int)) #<=================1 of Dictionary     
            for event in speech_event:
                    #Consider of the silence in the speech sample at the begining and end part
                    #So using ceil and floor
                osnet_frame = np.ceil(event['onset'] * audio_sr/hop_size).astype(int)
                offset_frame = np.floor(event['offset'] * audio_sr/hop_size).astype(int)
                frame_labels_vec[osnet_frame:offset_frame] = 1
            #labels.append(frame_labels_vec)
            
            #Calculate the probability of status transition If data contains speech 
            if sum(frame_labels_vec) != 0:
                for frame_index in range(len(frame_labels_vec)-1):
                    if frame_labels_vec[frame_index] == 0 and frame_labels_vec[frame_index+1] > frame_labels_vec[frame_index]:
                        previous_n = previous_n + 1
                        ns_temp = ns_temp + 1
                    elif frame_labels_vec[frame_index] == 0 and frame_labels_vec[frame_index+1] == frame_labels_vec[frame_index]:
                        previous_n = previous_n + 1
                        nn_temp = nn_temp + 1
                    elif frame_labels_vec[frame_index] == 1 and frame_labels_vec[frame_index+1] < frame_labels_vec[frame_index]:
                        previous_s = previous_s + 1
                        sn_temp = sn_temp + 1
                    elif frame_labels_vec[frame_index] == 1 and frame_labels_vec[frame_index+1] == frame_labels_vec[frame_index]:
                        previous_s = previous_s + 1
                        ss_temp = ss_temp + 1
                #tansition_amount = len(frame_labels_vec)-1
                #print('States Transition Probability:\n SS: {0:.6f} NS: {1:.6f}\n SN: {2:.6f} NN: {3:.6f}'.format(ss_temp/tansition_amount, ns_temp/tansition_amount, sn_temp/tansition_amount,nn_temp/tansition_amount))
                SS.append(ss_temp/previous_s)
                NN.append(nn_temp/previous_n)
                SN.append(sn_temp/previous_s)
                NS.append(ns_temp/previous_n)
            #Data==========================================================================
            #Short Time Energy
            #ste = ShortTimeEnergy(signal=audio_data,win_size=win_size,hop_size=hop_size)
                #ZCR
            #zcr = librosa.feature.zero_crossing_rate(y=audio_data,frame_length=win_size,hop_length=hop_size)
                #Spectral Centroid
            #cent = librosa.feature.spectral_centroid(y=audio_data, sr=sr,n_fft=win_size, hop_length=hop_size)
                #Spectral Entropy
            #entropy = Spectral_Entropy(y_Mag=audio_mag,sr=sr,n_short_blocks=10)
                #MFCCs and MFCCs with PCEN scaling
            #mfccs, pcens = PCEN_MFCC(y=audio_data,sr=sr,parameters=Parameters)
            mfccs = librosa.feature.mfcc(y=audio_data, sr=sr, n_mfcc=20, dct_type=2, norm='ortho')
            pcnns = PCNN(y=audio_data, sr=sr, parameters = Params)
            #Create Data Matrix
            temp_data = np.concatenate([#np.asarray([ste]).T, #<=================1 of Dictionary     
                                        #np.asarray([entropy]).T,
                                        #zcr.T,
                                        #cent.T,
                                        mfccs.T,
                                        pcens.T
                                                        #####################
                                                        #Other features here#
                                                        #####################
                                                        ],axis = 1)
            print(file_index)
            #convert list to np array
            temp_groups = (np.ones(len(frame_labels_vec))*file_index).astype(int)
            if file_index == 1:
                data_np = temp_data
                labels_np = frame_labels_vec
                groups_np = temp_groups
            else:
                data_np = np.concatenate([data_np,temp_data],axis = 0)
                labels_np = np.concatenate([labels_np,frame_labels_vec],axis = 0)
                groups_np = np.concatenate([groups_np,temp_groups],axis = 0)
            #For Group K Fold 
            #groups.append((np.ones(len(frame_labels_vec))*file_index).astype(int))
            #data.append(file_data)
    print('Data Size:\n{}'.format(data_np.shape))
    print('Labels Size: {}\n'.format(labels_np.shape))
    print('Groups Size: {}\n'.format(groups_np.shape))
    print('==================================================================')

    #Transition Matrix:
    trans_matrix = np.array([[np.mean(SS),1 - np.mean(SS)],[1 - np.mean(NN),np.mean(NN)]])
    #Saving the transition Matrix
    np.save(os.getcwd()+'/JPNotebookExported/' + Property + '_TransitionMatrix.npy', trans_matrix)
    print('Transition Matrix:\n ')
    print(trans_matrix)
    
    return data_np,labels_np,groups_np,trans_matrix
```

```python
Params = {
        'downsampling':22050,
        'win_size': 1024,
        'hop_size': 512,
        'min_freq': 80,
        'max_freq': 8000,
        'num_mel_filters': 128,
        'n_dct': 20}

#Train
Train_path = os.getcwd() + '/1_Dataset_Generate/audio/soundbanks/train/generated/Bimodal/'
TrainData,TrainLabel,TrainGroup,TrainTransMatrix = FeatureExtraction(Files_Dir = Train_path,
                                                          Parameters = Params,
                                                          Property = 'Train')

#Validate
Validate_path = os.getcwd() + '/1_Dataset_Generate/audio/soundbanks/Validate/generated/Bimodal/'
ValidateData,ValidateLabel,ValidateGroup,ValidateTransMatrix = FeatureExtraction(Files_Dir = Validate_path,
                                                                   Parameters = Params,
                                                                   Property = 'Validate')

#Test
Test_path = os.getcwd() + '/1_Dataset_Generate/audio/soundbanks/Test/generated/Bimodal/'
TestData,TestLabel,TestGroup,TestTransMatrix = FeatureExtraction(Files_Dir = Test_path,
                                                       Parameters = Params,
                                                       Property = 'Test')
```
